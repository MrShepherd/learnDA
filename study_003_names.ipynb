```python
years=range(1880,2011)
pieces=[]
columns=['name','sex','births']
for year in years:
    path='/home/shepherd/Work/git/pydata-book/ch02/names/yob%d.txt'%year
    frame=pd.read_csv(path,names=columns)
    frame['year']=year
    pieces.append(frame)
names_all=pd.concat(pieces,ignore_index=True)
```

```python
names_all
```

```python
total_births=names_all.pivot_table('births',index='year',columns='sex',aggfunc=sum)
total_births
```

```python
total_births.plot(title='gender situation')
```

```python
def add_prop(group):
    births=group.births.astype(float)
    group['prop']=births/births.sum()
    return group
names_all=names_all.groupby(['year','sex']).apply(add_prop)
names_all
```

```python
import numpy as np
np.allclose(names_all.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
pieces = []
for year, group in names_all.groupby(['year', 'sex']):
    pieces.append(group.sort_values(by='births', ascending=False)[:1000])
top1000 = pd.concat(pieces, ignore_index=True)
top1000.index = np.arange(len(top1000))
top1000
```

```python
boys=top1000[top1000.sex=='F']
girls=top1000[top1000.sex=='M']
total_births=top1000.pivot_table('births',index='year',columns='name',aggfunc=sum)
total_births
```

```python
subset=total_births[['John','Harry','Mary','Marilyn']]
subset.plot(subplots=True,figsize=(12,10),grid=False,title='number of births per year')
```

```python
import matplotlib.pyplot as plt
plt.figure()
table=top1000.pivot_table('prop',index='year',columns='sex',aggfunc=sum)
table.plot(title='sum of prop by year and sex',yticks=np.linspace(0,1.2,13),xticks=range(1880,2020,10))
```

```python
df=boys[boys.year==2010]
df
```

```python
prop_cumsum=df.sort_values(by='prop',ascending=False).prop.cumsum()
prop_cumsum[:10]
```

```python
prop_cumsum.searchsorted(0.5)
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.head()
```

```python
diversity.plot(title='number of popular names in top 50%')
```

```python
get_last_letter=lambda x:x[-1]
last_letters=names_all.name.map(get_last_letter)
last_letters.name='last_letter'
table=names_all.pivot_table('births',index=last_letters,columns=['sex','year'],aggfunc=sum)
subtable=table.reindex(columns=[1910,1960,2010],level='year')
```

```python
subtable.sum()
letter_prop=subtable/subtable.sum()
```

```python
all_names=top1000.name.unique()
mask=np.array(['lesl' in x.lower() for x in all_names])
lesley_like=all_names[mask]
lesley_like
```

```python
filtered=top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
table=filtered.pivot_table('births',index='year',columns='sex',aggfunc=sum)
table=table.div(table.sum(1),axis=0)
table.tail()
```

```python
table.plot(style={'M':'k-','F':'k--'})
```
